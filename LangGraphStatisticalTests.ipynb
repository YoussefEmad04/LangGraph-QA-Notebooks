```python
import numpy as np
from scipy.stats import pearsonr, ttest_ind, f_oneway, chi2_contingency, mannwhitneyu, kruskal

# Decorator to expose function to LLM
def llm_tool(func):
    func.is_tool = True
    return func

@llm_tool
def correlation_test(df: pd.DataFrame, col1: str, col2: str):
    """Calculate Pearson correlation between two numeric columns."""
    x = df[col1].dropna()
    y = df[col2].dropna()
    corr, p = pearsonr(x, y)
    return {"correlation": corr, "p_value": p}

@llm_tool
def independent_ttest(df: pd.DataFrame, col1: str, col2: str):
    """Perform independent t-test between two numeric columns."""
    x = df[col1].dropna()
    y = df[col2].dropna()
    t_stat, p = ttest_ind(x, y, equal_var=False)
    return {"t_stat": t_stat, "p_value": p}

@llm_tool
def anova_test(df: pd.DataFrame, *cols: str):
    """Perform one-way ANOVA across multiple numeric columns."""
    groups = [df[col].dropna() for col in cols]
    f_stat, p = f_oneway(*groups)
    return {"f_stat": f_stat, "p_value": p}

@llm_tool
def chi_squared_test(df: pd.DataFrame, col1: str, col2: str):
    """Perform Chi-squared test on two categorical columns."""
    contingency_table = pd.crosstab(df[col1], df[col2])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    return {"chi2_statistic": chi2, "p_value": p, "degrees_of_freedom": dof}

@llm_tool
def mann_whitney_u_test(df: pd.DataFrame, col1: str, col2: str):
    """Perform Mann-Whitney U test on two numeric columns."""
    x = df[col1].dropna()
    y = df[col2].dropna()
    u_stat, p = mannwhitneyu(x, y)
    return {"u_statistic": u_stat, "p_value": p}

@llm_tool
def kruskal_wallis_test(df: pd.DataFrame, *cols: str):
    """Perform Kruskal-Wallis H test across multiple numeric columns."""
    groups = [df[col].dropna() for col in cols]
    h_stat, p = kruskal(*groups)
    return {"h_statistic": h_stat, "p_value": p}
```

```python
from langgraph.graph import StateGraph, START, END, MessagesState
from langgraph.prebuilt import ToolNode
import google.generativeai as genai
from langchain_google_genai import ChatGoogleGenerativeAI
from google.colab import userdata
from IPython.display import display, Markdown

class State(MessagesState):
    pass

GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0, google_api_key=GOOGLE_API_KEY)

tools = [correlation_test, independent_ttest, anova_test, chi_squared_test, mann_whitney_u_test, kruskal_wallis_test]
tool_node = ToolNode(tools)

def llm_agent(state: State):
    schema_info = {}
    for col, dtype in df.dtypes.astype(str).to_dict().items():
      if dtype in ["int64", "float64"]:
        schema_info[col] = "numeric"
      elif dtype == "object" or "category" in dtype:
        schema_info[col] = "categorical"
      else:
        schema_info[col] = dtype

    system_prompt = {
        "role": "system",
        "content": f"""You are an expert data analysis assistant. Your primary function is to analyze the provided CSV data and perform statistical tests as requested by the user.

You can perform the following statistical tests:
- Pearson Correlation Test: To measure the linear relationship between two numeric variables.
- Independent Samples t-test: To compare the means of two independent groups for a numeric variable.
- One-Way ANOVA: To compare the means of three or more independent groups for a numeric variable.
- Chi-squared Test: To examine the relationship between two categorical variables.
- Mann-Whitney U Test: A non-parametric test to compare the distributions of two independent groups for a numeric variable.
- Kruskal-Wallis H Test: A non-parametric test to compare the distributions of three or more independent groups for a numeric variable.

When a user asks for an analysis or statistical test:
1. Identify the relevant columns in the dataframe based on their request and the available data.
2. Use the appropriate tool from your available tools to perform the analysis.
3. Present the results clearly, including the test statistic, p-value, and a brief interpretation of the results in the context of the data.

Important:
- Only numeric columns (int, float) should be used for tests requiring numerical input.
- Only categorical/object columns should be used for tests requiring categorical input.
- If a requested test cannot be performed due to incompatible column types, explain this clearly.

Here is the dataframe schema (column names and types):
{schema_info}"""
    }

    messages = [system_prompt] + state["messages"]

    # ✅ force output into proper LangChain AIMessage
    response = llm.invoke(messages)
    if isinstance(response, str):
        response = AIMessage(content=response)
    elif not hasattr(response, "content"):
        response = AIMessage(content=str(response))

    return {"messages": [response]}

workflow = StateGraph(state_schema=State)
workflow.add_node("agent", llm_agent)
workflow.add_node("tools", tool_node)
workflow.add_edge(START, "agent")

def route_agent(state):
    # Inspect the structure of the last message
    last_message = state["messages"][-1]
    print("Last message structure for routing:") # Debug info
    print(last_message)

    # Check if the message contains a tool call
    if hasattr(last_message, 'additional_kwargs') and 'tool_calls' in last_message.additional_kwargs and last_message.additional_kwargs['tool_calls']:
         return "tools"
    return END

workflow.add_conditional_edges(
    "agent",
    route_agent,
    {"tools": "tools", END: END}
)
workflow.add_edge("tools", "agent")

graph = workflow.compile()

def ask_data_question(question: str, chat_history=[]):
    try:
        result = graph.invoke({"messages": chat_history + [{"role": "user", "content": question}]})
        chat_history = result["messages"]
        answer = chat_history[-1].content
        display(Markdown(answer))
        return chat_history
    except Exception as e:
        print(f"An error occurred during graph execution: {e}")
        return chat_history


# ✅ Step 2: Add chatbot-like loop at the bottom
chat_history = []

print("📊 Data Analysis Assistant Ready! (type 'exit' to quit)\n")

while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        print("👋 Goodbye!")
        break

    chat_history = ask_data_question(user_input, chat_history)
```
